```python
def select(choice):
    video_path = 0
    if choice<=2:
        video_path = "video"+str(choice+1)+".mp4"
    if choice==3:
        video_path = 0

    Analyse(video_path)


    
```

```python
class MyWidget(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('Person Tracking Application')
        self.setGeometry(50, 50, 200, 150)
        
        font = QtGui.QFont()
        font.setPointSize(18)
        font.setBold(True)
        self.title = QLabel('Person Tracking', self)
        self.title.move(60, 50)
        self.title.setFont(font)
        self.title.setAlignment(Qt.AlignCenter)
        
        
        font_hint = QtGui.QFont()
        font_hint.setPointSize(12)
        self.hint = QLabel('Tutorial: This application allows you to choose 4 default videos in the system or upload your own video. It can identify and track the pedestrians in the video. You can also use camera to capture a real time video. If you want to quit the video, please press "q" in keyboard.', self)
        self.hint.move(110, 50)
        self.hint.resize(700,300)
        self.hint.setFont(font_hint)
        self.hint.setWordWrap(True)
        
        self.choose = QLabel('Make your choice:',self)
        self.choose.move(150,300)
        self.choose.setFont(font_hint)
        
        self.mycombobox = QComboBox(self)
        self.mycombobox.addItems(['video1', 'video2', 'video3', 'Camera'])
        self.mycombobox.setCurrentIndex(1)
        self.mycombobox.move(400,300)
        self.mycombobox.setFont(font_hint)
        
        self.upload = QPushButton('Upload' ,self)
        self.upload.move(650,500)
        self.upload.clicked.connect(self.clickButtonUpload)

        self.button = QPushButton('Quit', self)
        self.button.move(800,500)
        self.button.clicked.connect(self.clickButtonCloseWindow)

        self.go = QPushButton('Analyse' ,self)
        self.go.move(650,300)
        self.go.clicked.connect(self.clickButtonGoAnalyse)
        
    def clickButtonCloseWindow(self):
        self.close()
        
    def clickButtonGoAnalyse(self):
        choice = self.mycombobox.currentIndex()
        select(choice)
        
    def clickButtonUpload(self):
        directory = PyQt5.QtWidgets.QFileDialog.getOpenFileName(self,  "选取文件","./", "All Files (*);;Text Files (*.txt)") 
        video_path = directory[0]
        Analyse(video_path)

        
```

```python
def show():
    app = QApplication(sys.argv)
    
    
    w = MyWidget()
    w.resize(1000, 600)
    w.move(300, 300)
    w.show()
    
    
    sys.exit(app.exec_())
```

```python
def Analyse(video_path):
    cap = cv2.VideoCapture(video_path)      
    pts = []
    while cap.isOpened():
        # Read a frame from the video
        success, frame = cap.read()
        
        
        if success:
            # Run YOLOv8 inference on the frame
            results = model.track(frame,classes=[0],tracker="bytetrack.yaml")
            #track = model.track(frame)
            
            
            get_center = results[0].boxes.data.numpy()
            
            for i in range(0,len(get_center)):
                center = (math.ceil((get_center[i][0]+get_center[i][2])/2),math.ceil((get_center[i][1]+get_center[i][3])/2))
                pts.append(center)
            
            
        
            # Visualize the results on the frame
            
            
                
            annotated_frame = results[0].plot()
            
            
            # Display the annotated frame
            for j in range(0, len(pts)):
                print(pts[j])
                cv2.circle(annotated_frame,(pts[j]),1,(0, 255, 255),8)   
            text = "People number: "+ str(len(get_center))
            cv2.putText(annotated_frame, text, (40, 50), cv2.FONT_HERSHEY_PLAIN, 2.0, (0, 255, 255), 3)  
            cv2.imshow("Person tracking", annotated_frame)
            
            # Break the loop if 'q' is pressed
            if cv2.waitKey(1) & 0xFF == ord("q"):
                break
        else:
            # Break the loop if the end of the video is reached
            break

    # Release the video capture object and close the display window
    cap.release()
    cv2.destroyAllWindows()
```
